```python
import pandas as pd
import numpy as nm

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv'
data_set = pd.read_csv(path)
data_set.head()
```

```python
data_set = data_set.rename(columns={'math score':'math_score', 'reading score':'reading_score', 'writing score':'writing_score'})
```

```python
data_set.math_score.hist()
```

```python
data_set.plot.scatter(x='math_score', y='reading_score')
```

```python
# default fit_reg=True -> we have regration lines; if we use fit_reg=False we will have plot without regration lines
sns_plot = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=data_set)  
```

```python
# set name of axises
sns_plot.set_xlabels('Math score')
sns_plot.set_ylabels('Reading score')
plt.show()
```

```python
path = 'https://stepik.org/media/attachments/course/4852/iris.csv'
df = pd.read_csv(path, index_col=0)
df.sort_values('species', ascending=False).head()
```

```python
for column in list(df):
    sns.distplot(df[column], kde_kws={'label':column})
    
```

```python
sns.violinplot(y=df['petal length'])
```

```python
sns.pairplot(df,  hue="species" )
```

```python
df.corr()
```

```python
#use df.iloc[0:-1,0:-1] for build heatmap without 'species'
sns.heatmap(df.iloc[0:-1,0:-1].corr(),\
            annot=True, cmap= 'coolwarm')
```
