```python
import pandas as pd
import numpy as np
```

```python
path = 'https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv'
data_set = pd.read_csv(path)
data_set.head()
```

```python
mean_scores = data_set.groupby(['race/ethnicity', 'gender'], as_index=False) \
  .aggregate({'math score':'mean', 'reading score':'mean','writing score':'mean'}) \
  .rename(columns={'math score':'mean_math_score', 'reading score':'mean_reading_score', 'writing score':'mean_writing_score'})
```

# one more way to aggregate by mean() for all numeric features + rename

```python
data_set.groupby('gender', as_index=False).mean() \
  .rename(columns={'math score':'mean_math_score', 'reading score':'mean_reading_score', 'writing score':'mean_writing_score'})
```
