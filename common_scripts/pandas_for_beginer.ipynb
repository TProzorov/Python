# Find MEAN and VARIANCE (DISPERSION) group by column ('lunch')

```python
data_set.groupby('lunch').agg({'math score':'mean','reading score':'mean','writing score':'mean'})
```

```python
data_set.groupby('lunch').agg({'math score':'var','reading score':'var','writing score':'var'})
```

```python
data_set.groupby('lunch').describe()
```

# rename

```python
data_set = data_set.rename(columns = {'parental level of education':'parental_level_of_education', 'test preparation course':'test_preparation_course','math score':'math_score','reading score':'reading_score','writing score':'writing_score'})
```

```python
data_set.head()
```

# Use QUERY

```python
data_set.query("writing_score > 74 & gender == 'male'")
```

# with QUERY we must use @ with variables

```python
our_variable = 74
```

```python
data_set.query("writing_score > @our_variable & gender == 'male'")
```
