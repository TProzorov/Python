```python
import pandas as pd
import numpy as np
```

```python
path = 'https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv'
data_set = pd.read_csv(path)
data_set.head()
```

```python
data_set=data_set.rename(columns={'math score':'math_score', 'reading score':'reading_score', 'writing score':'writing_score'})
```

# add new one column

```python
data_set['total_score'] = data_set.math_score + data_set.reading_score + data_set.writing_score
data_set.head()
```

# add more then one column at once

```python
new_data_set = data_set.assign(total_score_log = np.log(data_set.total_score), \
                               mean_score = data_set.total_score/3)
new_data_set.head()
```

# delete or drop column (axis = 1) or row (axis = 0)

```python
new_data_set.drop(['total_score'], axis=1)
```
